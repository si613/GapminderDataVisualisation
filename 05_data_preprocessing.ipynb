## Investigating the shape and values from the 4 datasets

```python
print('Meta Country Data:',country.shape)
print('Population Data:',population.shape)
print('Fertility Rate Data:',fertility.shape)
print('Life Expectancy Data:',l_expectancy.shape)
```

There is an extra entry in three datasets. <br> <br>
We will convert the country names into sets and apply set functions to see what values are missed.

```python
countries_country = set(country['CountryName'])
countries_population = set(population['Country Name'])
countries_fertility = set(fertility['Country Name'])
countries_expectancy = set(l_expectancy['Country Name'])
```

Countries that are in the meta data but not in the other three:

```python
countries_country - (countries_population | countries_fertility | countries_expectancy)
```

```python
population[population['Country Name'] == "Côte d'Ivoire"]
```

```python
population[population['Country Name'] == "Cote d'Ivoire"]
```

Above, we discover that there are diacritical marks applied on the names of countries in the meta data that are missing in the other three datasets.

```python
population[population['Country Name'] ==  "Korea, Dem. People's Rep."]
```

```python
population[population['Country Name'] == "Korea, Dem. People’s Rep."]
```

Above, there is a difference in the apostrophe used in the word "People's" in the two datasets

```python
population[population['Country Name'] ==  "East Asia & Pacific (IDA & IBRD)"]
```

```python
population[population['Country Name'] ==  "East Asia & Pacific"]
```

Above, we see that the text in the brackets is present in the meta dataset, but not in the other datasets

The investigation from the above three examples reveal that while the 9 countries are present all 4 datasets, the differences in spelling the country names cause them to not be matched with the meta dataset. <br> <br>
Thus, we must rely on the country code to merge the diffrent datasets.

Now, let's look at countries that are not in the meta dataset, but in the other 3:

```python
(countries_population | countries_fertility | countries_expectancy) - countries_country
```

Ignoring the 9 countries that are spelt/written differently, we find "Not classified" to be an anomaly that must be removed

```python
country[country['CountryName'] == 'Not classified']
```

```python
population[population['Country Name'] == "Not classified"]
```

```python
# dropping not classified from all the dataframes

population = population[population['Country Name'] != 'Not classified']
fertility = fertility[fertility['Country Name'] != 'Not classified']
l_expectancy = l_expectancy[l_expectancy['Country Name'] != 'Not classified']

print('Meta Country Data:',country.shape)
print('Population Data:', population.shape)
print('Fertility Rate Data:', fertility.shape)
print('Life Expectancy Data:', l_expectancy.shape)
```

Now, the shape of the dataframe is equal!

## Melting the Datasets
Melting the dataset unpivots them and reshapes it

Melting the population dataset:- 

```python
population_melt = pd.melt(population, id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Population')

population_melt
```

```python
# sorting by country name and year
population_melt = population_melt.sort_values(by=['Country Name', 'Year'])

# resetting the index
population_melt = population_melt.reset_index(drop=True)

population_melt.head()
```

```python
population_melt.info()
```

```python
# converting Year column to integer
population_melt['Year'] = population_melt['Year'].astype(int)

population_melt.info()
```

Melting the fertility rate dataset:-

```python
fertility_melt = pd.melt(fertility, id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Fertility Rate')

# converting the Year column to integer
fertility_melt['Year'] = fertility_melt['Year'].astype(int)

# sorting by Country Name and Year
fertility_melt = fertility_melt.sort_values(by=['Country Name', 'Year'])

# resetting the index
fertility_melt = fertility_melt.reset_index(drop=True)

fertility_melt
```

Melting the life expectancy dataset:-

```python
expectancy_melt = pd.melt(l_expectancy, id_vars=['Country Name', 'Country Code'], var_name='Year', value_name='Life Expectancy')

# converting Year column to integer
expectancy_melt['Year'] = expectancy_melt['Year'].astype(int)

# sorting by Country Name and Year
expectancy_melt = expectancy_melt.sort_values(by=['Country Name', 'Year'])

# resetting the index
expectancy_melt = expectancy_melt.reset_index(drop=True)

expectancy_melt
```

________________________________

## Merging the Datasets

```python
# merging population_melt and fertility_melt

data_melt = pd.merge(population_melt, fertility_melt, on=['Country Name', 'Country Code', 'Year'], how='inner')

data_melt
```

```python
# merging data_melt with expectancy_melt

data_melt = pd.merge(data_melt, expectancy_melt, on=['Country Name', 'Country Code', 'Year'], how='inner')

data_melt
```

Checking the country meta data to merge with data_melt
